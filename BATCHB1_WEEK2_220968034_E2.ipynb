## Using the given BENGALURU HOUSE PRICES DATASET, perform data preprocessing and answer the following questions.

```python
import pandas as pd
df=pd.read_csv("Bengaluru_House_Prices.csv")
df
```

### 1. Compute the total number of missing values in the DataFrame

```python
total_NaN_count=df.isna().sum().sum()
print("Total number of missing values in the dataframe is: ",total_NaN_count)
```

### 2.Compute the percentage of missing values in the DataFrame. If a column has more than 15% missing values then drop (or remove) the column from the DataFrame.

```python
for i in df.columns:
    missing_values=df[i].isna().sum()
    missing_percent=(missing_values/13319)*100    
    if(missing_percent>15.):
        df = df.drop(columns=[i])
df
```

### 3. Which column has the most number of missing values?

```python
present=0
for i in df.columns:
    missing=df[i].isna().sum()
    if(missing>=present):
        present=missing
        index=i
print("column with highest missing values is: ",index)
```

### 4. Drop (or remove) the rows containing the missing values in the location, size, and total_sqft columns

```python
columns=["location","size","total_sqft"]
for i in columns:
    df = df[df[i].notna()]
df
```

### 5.drop all the rows in the bath column containing more than 5 bathrooms.

```python
df=df[df["bath"].notna()]
mask=df["bath"]<=5
df=df[mask]
df
```

### 6. Find the houses available in each area

```python
houses_by_area = df.groupby(['area_type','location']).size()
houses_by_area
```

### 7. Find the top five areas where the large number of houses available

```python
top_five_areas=houses_by_area.sort_values(ascending=False).head(5)
print("The top five areas are:")
top_five_areas
```

### 8. Visualize the house price based on build up area of top five locations where the large number of houses available (Visualize by independent charts.)

```python
import matplotlib.pyplot as plt
houses_by_location = df['location'].value_counts()
top_five_locations = houses_by_location.head(5).index
top_five_data = df[df['location'].isin(top_five_locations)]
for location in top_five_locations:
    location_data = top_five_data[top_five_data['location'] == location]
    plt.figure(figsize=(10, 6))
    plt.scatter(location_data['total_sqft'], location_data['price'])
    plt.title(f'House Price vs Build-up Area in {location}')
    plt.xlabel('Build-up Area (sqft)')
    plt.ylabel('House Price')
    plt.grid()
    plt.show()
```
